# Задание 1. Первичный анализ данных с Pandas

### [1]Импортируем необходимые библиотеки:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar, minimize, Bounds
%matplotlib inline
```

### Считываем данные по росту и весу из файла weights_heights.csv в объект Pandas DataFrame:

```python
data = pd.read_csv('weights_heights.csv', index_col='Index')
print(data)
```

### Строим гистограмму распределения роста подростков из выборки data:

```python
data.plot(y='Height', kind='hist', color='red',  title='Height (inch.) distribution')
```

### [2]Посмотрим на первые 5 записей в data: 

```python
data.head()
```

### Строим гистограмму распределения веса подростков из выборки data:

```python
data.plot(y='Weight', kind='hist', color='green',  title='Weight (pound) distribution')
```

### Создаем признак индекс массы тела и добавляем его в data:

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2

data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], row['Weight']), axis=1)
print(data)
```

### [3]Построим картинку, на которой будут отображены попарные зависимости признаков , 'Height', 'Weight' и 'BMI' друг от друга:

```python
sns.pairplot(data)
```

### [4]Создаем в DataFrame data новый признак weight_category и строим boxplot:

```python
def weight_category(weight):
    if weight < 120:
        weight_cat = 1
    elif weight >= 150:
        weight_cat = 3
    else: weight_cat = 2
    return weight_cat
    
data['weight_cat'] = data['Weight'].apply(weight_category)
print(data)
```

```python
box_plot = sns.boxplot(y=data["Height"], x=data["weight_cat"])
box_plot.set_xlabel('Весовая категория')
box_plot.set_ylabel('Рост')
```

### [5]Строим scatter plot зависимости роста от веса:

```python
data.plot(x='Weight', y='Height', kind='scatter', color='blue',  title='Dependence of height on weight')
```

# Задание 2. Минимизация квадратичной ошибки

### [6]Создаем функцию, которая по двум параметрам  w0  и  w1  вычисляет квадратичную ошибку приближения зависимости роста  y  от веса  x  прямой линией  y=w0+w1∗x :

```python
def error(w):
    Error = 0
    n = data.shape[0]
    for i in range(1,n+1):
        Error = Error + (data['Height'][i] - (w[0] + w[1] * data['Weight'][i])) ** 2
    return Error
```

### [7]Построим на графике из п. 5 Задания 1 две прямые, соответствующие значениям параметров  (w0,w1)=(60,0.05)  и (w0,w1)=(50,0.16):

```python
def plot_line(w0, w1, color, label):
    min_Weight = data['Weight'].min()
    max_Weight = data['Weight'].max()
    x = np.linspace(int( min_Weight),int(max_Weight),10)
    y = w0 + w1 * x
    plt.plot(x, y, color=color, label=label)
    plt.legend()
    return

data.plot(x='Weight', y='Height', kind='scatter', color='blue',  title='Dependence of height on weight')
plot_line(60,0.05,'red', 'y=60+0.05*x')
plot_line(50,0.16,'yellow', 'y=50+0.16*x')
```

### [8]Построим график зависимости функции ошибки, посчитанной в п. 6, от параметра  $w_1$  при  $w_0  = 50$:

```python
x = np.linspace(-5,5,100)
y = error([50, x])
plt.plot(x,y)
plt.xlabel('w1')
plt.ylabel('error')
plt.title('Dependence of error on w1')
```

### [9]Методом оптимизации найдем "оптимальный" наклон прямой, приближающей зависимость роста от веса, при фиксированном коэффициенте  $w_0=50$:

```python
w1_opt = minimize_scalar((lambda x: error([50, x])),bounds=(-5, 5), method='Bounded').x
w1_opt
```

### Строим оптимальную прямую:

```python
data.plot(x='Weight', y='Height', kind='scatter', color='blue',  title='Dependence of height on weight')
plot_line(50,w1_opt,'red','y=50+'+str(round(w1_opt,3))+'*x')
```

### [10]Построим 3D-график зависимости функции ошибки, посчитанной в п.6 от параметров  $w_0$  и $w_1$:

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
w0 = np.arange(-300, 400, 10)
w1 = np.arange(-5, 5, 0.25)
w0, w1 = np.meshgrid(w0, w1)
Z = error([w0,w1])
surf = ax.plot_surface(w0, w1, Z)
ax.set_xlabel('Intercept')
ax.set_ylabel('Slope')
ax.set_zlabel('Error')
plt.show()
```

### [11]Найдем минимум квадратичной ошибки по параметрам $w_0$ и $w_1$:

```python
w_opt = minimize(error, [0,0], method='L-BFGS-B', bounds = Bounds([-100,-5],[100,5])).x
w_opt
```

### Построим на графике из п. 5 Задания 1 прямую, соответствующую найденным оптимальным значениям параметров  $w_0$  и  $w_1$

```python
data.plot(x='Weight', y='Height', kind='scatter', color='blue',  title='Dependence of height on weight')
plot_line(w_opt[0],w_opt[1],'red','y='+str(round(w_opt[0],3))+'+'+str(round(w_opt[1],3))+'*x')
```
